```python
# Bases de datos
mass = pd.read_csv("Mass.csv")
mass[['1','0','1n']]
mass
```

```python
matrix = []
lista = []

for i in range(0, 5842):
    for j in mass.columns:
        
        # Se carga la entrada en una variable
        entry = mass.loc[i,j]
        
        # Se almacenan los datos diferentes de NaN
        if (not mass.iloc[[i]][j].isna().bool()):
            lista.append(entry)
            
        # Solo interesan los primeros datos
        if len(lista) == 5:
            break
                
    # Se almacenan los datos en una matriz
    matrix.append(lista)
    lista = []
    
for i in range(len(matrix)):

    # Se divide la entrada por 10
    matrix[i][1] = int(matrix[i][1]/10)
    
    # Se extraen los datos de interÃ©s
    if (matrix[i][3] == 'i' or matrix[i][3] == 'j' or matrix[i][3] == 'm' or matrix[i][3] == 'p' or matrix[i][3] == 'n' or matrix[i][3] == 'q' or matrix[i][3] == 'r' or matrix[i][3] == 'x'):
        matrix[i] = [matrix[i][0], matrix[i][1], matrix[i][2], matrix[i][4]]
    else:
        matrix[i] = [matrix[i][0], matrix[i][1], matrix[i][2], matrix[i][3]]
matrix
```
